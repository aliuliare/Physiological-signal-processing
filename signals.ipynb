# BIOELECTRICAL AND PHYSIOLOGICAL SIGNAL PROCESSING


## **Bioelectrical Signals**
1. **Electroencephalogram (EEG)** – Reflects the electrical activity of the brain.
2. **Electrocorticogram (ECoG)** – An invasive recording of brain activity from the uncovered brain surface.
3. **Evoked Potentials (EPs)** – Brain activity triggered by sensory stimuli (visual, auditory).
4. **Electrocardiogram (ECG)** – Electrical activity of the heart.
5. **Electrogram (EG)** – Intracardiac recording used in pacemakers and defibrillators.
6. **Electromyogram (EMG)** – Electrical activity of skeletal muscles.
7. **Electroneurogram (ENG)** – Measures nerve conduction velocity.
8. **Electroretinogram (ERG)** – Electrical activity of the retina in response to light stimuli.
9. **Electrooculogram (EOG)** – Records the steady corneal-retinal potential related to eye movements.
10. **Electrogastrogram (EGG)** – Electrical signals from stomach muscles controlling contractions.

## **Other Physiological Signals**
11. **Blood Pressure Signal** – Often recorded in multimodal studies, it reflects the cardiovascular state.
12. **Respiration Signals (Nasal and Abdominal Respiration)** – Monitors breathing patterns, used in sleep disorder studies.
13. **Heart Rate Variability (HRV)** – Derived from ECG and provides information on autonomic nervous system function.

# ECG SIGNAL

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "data/mitbih_test.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info(), df.head()

# Select the first row as an example heartbeat signal
heartbeat_signal = df.iloc[0, :].values

# Plot the heartbeat signal
plt.figure(figsize=(10, 4))
plt.plot(heartbeat_signal, label="Heartbeat Signal")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Heartbeat Signal Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot multiple heartbeat signals
plt.figure(figsize=(12, 6))

num_signals = 5  # Number of rows to plot
for i in range(num_signals):
    plt.plot(df.iloc[i, :].values, label=f"Heartbeat {i+1}")

plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Multiple Heartbeat Signals")
plt.legend()
plt.grid(True)
plt.show()
```
